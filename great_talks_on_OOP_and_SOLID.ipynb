* Should square inherit from rectangle? why? why not?
* How do dynamic languages stack up against statically typed ones? (which effort more flexiblity and at what potential cost?)

This episode also touches on why certain concepts are harder to relate to if your background is only in dynamic langauges. For instance, dynamic langauges embrace the interface segregation pattern in its most extreme form - an object only 'implements' an interface consisting of the methods it calls on another object! This can make claims on SOLID quite elusive to programmers not familiar with static typing.

As for the Interface Segragation Pattern, there is still lesson to be learned there even if you code in Ruby or Python. For instance, here is a very nice [blog post](http://rubyblog.pro/2017/07/solid-interface-segregation-principle) that reads into the ISP in the context of Ruby.

* What does Bob Martin, a person advocating small methods with descriptive names, think about dense languages (where APL is on the far end of dense but Ruby still being more dense then most)? Big surprise and a great lesson on code readability and quality regarding of context.

A big takeaway here is the importance of tests. Over last couple of years I went from not knowing what tests where, seeing them as necessary evil, thinking they might be useful maybe in some context to now not imagining a world without having them.

Again, there are some caveats to this. For instance, I think if you are working on an ML project and have an end to end way of running it (even on subset of data), have some performance baseline and do not mutate too much state along the way, you can be perfectly fine without tests. But once you start writing software that has to do more than one thing, without tests, you are screwed.

Another Hanselminutes [podcast episode](https://hanselminutes.com/150/uncle-bob-martin-solid-this-time-with-feeling) with Bob Martin where they discuss the falling out between Joel Spolsky and Bob Marting regarding TDD.

I do not agree with everything that has been said in this podcast.

First of all, you have various types of software. A pacemaker software should be developed to a different level of quality than the site for a local newspaper requires.

Secondly, writing software has a built in feedback loop that can come at no to little cost. You work on a project, you learn from it. As simple as this. Initially, you only risk wasting your time and nothing else. When you practice medicine, you risk the life of a patient.

Nonetheless, it is probably great to have a mentor and learn on the job. And it is very likely that Bob Martin is spot on when it comes to writing enterprise software in a corporate setting.

Nonetheless, despite not agreeing with some statements being made as absolutes, this was still a good listen.

> `QFT`: architecture comes from tests is horseshit!!!

Three laws of TDD by Bob Martin:
1. You are not allowed to write a single line of production code untill you write a failing unit test 
2. You do not write more of a unit test than is sufficient to fail (non compiling is failing) so you can't write a lot of unit tests until you must write production code.
3. You can't write more production code than is sufficient to pass currently failing test (you can't write a failing unit test and then run of to write a bunch of production code).
